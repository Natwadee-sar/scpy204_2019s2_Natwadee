```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy as astp
```

```python
table = pd.read_csv('./skyserverdata.csv')
table
```

```python
#Choose specific column from the table

z_info = table['z']
z_err = table['zErr']

pm_info = table['petroMag_g']
pm_err = table['petroMagErr_g']

print(z_info)
```

```python
print(z_err)
```

```python
print(pm_info)
```

```python
print(pm_err)
```

```python
#draw graph
x = z_info
x_err = z_err<=1
y = pm_info
y_err = pm_err<=1

figure = plt.figure(0)
axes = plt.gca()
axes.set_xlim([0.3,0.4])
axes.set_ylim([12,29])
plt.plot(x, y, '.', markersize=0.8, color='red')
plt.errorbar(x, y, xerr=z_err, yerr=pm_err, fmt='o', markersize=0.7, alpha=0.5, elinewidth=0.5, color='black')
plt.title('Relation between Petrosian Magnitude and Final Red Shift',loc='center', )
plt.ylabel('Petrosian Magnitude')
plt.xlabel('Final Red Shift (z)')
plt.show()
```
